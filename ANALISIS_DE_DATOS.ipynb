##### 5. ANÁLISIS ESTADÍSTICO ANOVA
---

```python
df['GRD -Código'] = df['GRD -Código'].astype(str)

label_enc = LabelEncoder()

excluded_cols = ['GRD -Código', 'GRD -Descripción', 'Tipo GRD']

candidate_cols = [col for col in df.columns if col not in excluded_cols and df[col].nunique() > 1]

anova_results = []


for col in candidate_cols:
    try:

        df_copy = df[[col, 'GRD -Código']].dropna()

        # categórica
        if df_copy[col].dtype == 'object':
            df_copy[col] = label_enc.fit_transform(df_copy[col])

        groups = [df_copy[df_copy['GRD -Código'] == grd][col] for grd in df_copy['GRD -Código'].unique()]

        # Ejecutar ANOVA solo si hay al menos 2 grupos con más de 1 valor
        valid_groups = [g for g in groups if len(g) > 1]
        if len(valid_groups) > 1:
            stat, p_value = f_oneway(*valid_groups)
            anova_results.append((col, stat, p_value))
    except Exception as e:
        print(f"Error procesando {col}: {e}")

anova_df = pd.DataFrame(anova_results, columns=['Variable', 'F-statistic', 'P-value'])
anova_df.sort_values('P-value', inplace=True)

print(" Variables con alta asociación al GRD -Código (p < 0.05):\n")
print(anova_df[anova_df['P-value'] < 0.05])

anova_df.to_csv("/content/anova_grd_codigo_codificado.csv", index=False)
```

Al utilizar las variables estadísticamente significativas, es decir aquellas con un p value menor a 0,05 de cada variable de entrada en relación a la variable de salida ***GRD -Código*** se obtuvo un mejor rendimiento del modelo, reflejado en la métrica de accuracy. Esto debido a que estamos eliminando factores irrelevantes en la predicción como lo son las variables; *Fecha de ingreso, Fecha de egreso, Proc4 - Proc15, Proc17 - Proc27, Dxr 10.* Una vez realizado esto eliminamos aquellas variables que no proporciona sufieciente información relevante en la predicción, evitando así un mayor gasto computacional, como es el caso de la variable *Num caso*

Generación de excel para obtener etiquetas con tipografía adecuada para su posterior implementación en una página web monolítica

```python
import pandas as pd
import numpy as np

variables = [
'Edad','Grupo Edad','Días estancia','Dx principal de egreso ','Cuidados intensivos','Días de Unidad Cuidado Intensivo',
        'Tipo de ingreso','Dx principal de egreso .1','Código causa externa','Proc1','Proc2','Proc3','Dxr 1','Dxr 2','Dx Ppal 3 Caracteres',
        'Infección Quirurgica','Tipo servicio','Infecciones','Dx de ingreso','ServicioAlta','Proc4','Sexo','Dxr 3','Causa externa',
        'Aseguradora -Código-','Proc5','Aseguradora -Descripción-','Dxr 4','Causa Basica de muerte','Dxr 5','Codigo de ciudad','Proc6','Dxr-6',
        'Proc7','Dxr 7','Proc8','Proc28','Dxr 8','Dxr 9','Proc9','Proc16','Situacion al alta','GRD -Código','GRD -Descripción'
]

clases_por_variable = {}
max_len = 0

for var in variables:
    if var in df_selected.columns:
        unique_vals = df_selected[var].dropna().unique()
        cleaned_vals = []
        for val in unique_vals:
            if isinstance(val, (np.integer, np.int64, np.int32)):
                cleaned_vals.append(int(val))
            elif isinstance(val, (np.floating, np.float64, np.float32)):
                cleaned_vals.append(float(val))
            else:
                cleaned_vals.append(str(val))
        sorted_vals = sorted(cleaned_vals, key=lambda x: str(x))
        clases_por_variable[var] = sorted_vals
        max_len = max(max_len, len(sorted_vals))

data_dict = {}
for var, valores in clases_por_variable.items():
    data_dict[var] = valores + [None] * (max_len - len(valores))

df_clases_unificado = pd.DataFrame(data_dict)

df_clases_unificado.to_excel("/content/drive/MyDrive/Info/clases_por_variable.xlsx", index=False)

print("✅ Archivo 'clases_por_variable_unificado.xlsx' creado con éxito en una sola hoja.")
```
